```python

import backoff
import openai
openai.api_key = 'KEY'
```

```python

def run_gpt4(prompt="Hello! What's the capital of China?", n=1, oa_model_type='gpt-4', max_tokens=None):
    if max_tokens is None:
        completion = openai.ChatCompletion.create(model=oa_model_type,
                                                  n=n,
                                                  temperature=0.9,
                                                  messages=[
                                                      {"role": "system", "content": prompt}
                                                  ])
    else:
        completion = openai.ChatCompletion.create(model=oa_model_type,
                                                  n=n,
                                                  temperature=0.9,
                                                  max_tokens=max_tokens,
                                                  messages=[
                                                      {"role": "system", "content": prompt}
                                                  ])

    #print(f"calling openai with params: {(oa_model_type, n, 0.9)}")


    to_ret = []

    for c in completion['choices']:
        to_ret.append(c['message']['content'])
    return to_ret
```

```python

@backoff.on_exception(backoff.expo, openai.error.RateLimitError,
                      
                     max_tries=10,
                      raise_on_giveup=False,)
def run_gpt4_backoff(*args,**kwargs):
    return run_gpt4(*args,**kwargs)
```

```python
run_gpt4_backoff('中国的首都是哪里？')[0]
```
